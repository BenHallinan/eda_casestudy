Comparison of Public vs Private Germany,China,United States

```python
countries_of_interest = ['China', 'Germany', 'United States']
```

```python
indicators_of_interest = [
    'Health expenditure per capita, PPP',
    'Health expenditure, private (% of GDP)',
    'Health expenditure, private (% of total health expenditure)',
    'Health expenditure, public (% of GDP)',
    'Health expenditure, public (% of government expenditure)',
    'Health expenditure, public (% of total health expenditure)',
    'Health expenditure, total (% of GDP)',
    'Health expenditure, total (current US$)'
]
```

```python
filtered_df = df[(df['Country Name'].isin(countries_of_interest)) & 
                 (df['Indicator Name'].isin(indicators_of_interest))]
```

```python
year_columns = [str(year) for year in range(1960, 2016)]
melted_df = filtered_df.melt(id_vars=['Country Name', 'Indicator Name'], 
                             value_vars=year_columns, 
                             var_name='Year', 
                             value_name='Value')
```

```python
melted_df.head()
```

```python
health_per_capita_ppp = 'Health expenditure per capita, PPP'
subset = melted_df[melted_df['Indicator Name'] == health_per_capita_ppp]
```

```python
plt.figure(figsize=(15, 7))
for country in ['China', 'Germany', 'United States']:
    country_data = subset[subset['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['Value'],label= country)
plt.title(health_per_capita_ppp)
plt.xticks(rotation=45)
plt.ylabel('Expenditure Value')
plt.legend(title='Country')
plt.xlabel('Year')

plt.show()
```

```python
indicator_name = "Life expectancy at birth, total (years)"
life_expectancy_data = df[df['Indicator Name'] == indicator_name]
```

```python

plt.figure(figsize=(15, 7))

for country in ['China', 'Germany', 'United States']:
    country_data = life_expectancy_data[life_expectancy_data['Country Name'] == country]
    plt.plot(country_data.columns[4:], country_data.iloc[0, 4:], label=country)  

plt.title("Yearly Life Expectancy (1960-2015)")
plt.xticks(rotation=45)
plt.ylabel('Life Expectancy at Birth (Years)')
plt.xlabel('Year')
plt.legend(title='Country')

plt.tight_layout()
plt.show()
```
